## 4. Batalha 1 — A tirania dos tipos de dados

Nos slides, vimos que datas não podem ser tratadas como texto e que números em formato string quebram análises. fileciteturn2file0

### Tarefa
Converta corretamente, quando fizer sentido:
- `data`
- `receita`
- `lucro`
- `quantidade`

### Orientação
- Investigue valores estranhos antes de converter
- Pense no uso de `errors='coerce'`
- Registre em markdown ou comentário quais problemas encontrou


```python
# Explore valores problemáticos e faça as conversões de tipo
# Ver exemplos das colunas problemáticas
df['data'].head(10)
df['receita'].head(10)
df['lucro'].head(10)
df['quantidade'].head(10)

# Ver valores únicos (ajuda a detectar sujeira)
df['receita'].unique()
df['lucro'].unique()

# Converter DATA
df['data'] = pd.to_datetime(df['data'], errors='coerce')


# Converter RECEITA (removendo R$, espaços e vírgulas)
df['receita'] = (
    df['receita']
    .astype(str)
    .str.replace('R$', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.strip()
)

df['receita'] = pd.to_numeric(df['receita'], errors='coerce')


# Converter LUCRO (garantir numérico)
df['lucro'] = pd.to_numeric(df['lucro'], errors='coerce')


# Converter QUANTIDADE (garantir inteiro)
df['quantidade'] = pd.to_numeric(df['quantidade'], errors='coerce')

df.info()
df.isna().sum()
```

### Reflexão curta
Explique:
1. Por que deixar `data` como texto pode quebrar análises temporais?

Quando a coluna data está como texto (object), o pandas não reconhece os valores como datas reais. Isso impede operações típicas de análise temporal. Por exemplo, a ordenação passa a ser alfabética e não cronológica, o que pode colocar datas fora de ordem. Além disso, não é possível extrair informações como mês, ano ou dia com .dt, nem calcular diferenças entre datas. Como consequência, análises como tendência ao longo do tempo, sazonalidade ou agrupamentos mensais ficam incorretas ou inviáveis.

2. Por que deixar `receita` como string pode distorcer cálculos?

Quando a coluna receita está como string, os valores não são tratados como números. Isso impede cálculos matemáticos corretos. Operações como soma e média podem falhar ou gerar resultados errados, pois o Python pode interpretar os valores como texto e realizar concatenação em vez de soma. Além disso, a presença de símbolos como "R$", vírgulas ou espaços impede a conversão automática para número. A ordenação também fica incorreta, pois segue regra alfabética. Isso compromete diretamente indicadores como receita total, média e lucro, tornando a análise pouco confiável.

## 6. Batalha 3 — O ataque dos clones (duplicidades)

A aula alerta que nem toda duplicidade é erro automático: pode haver compra legítima repetida ou dupla inserção do sistema. fileciteturn2file0

### Tarefa
1. Investigue se há linhas duplicadas
2. Analise se a duplicidade parece nociva para os KPIs
3. Escolha uma estratégia:
   - remover duplicidades totais?
   - remover apenas com base em certas colunas?
   - manter?
4. Justifique a decisão


```python
# Investigue duplicidades e trate o problema conscientemente
# Verificar duplicatas completas
df.duplicated().sum()

# Visualizar duplicatas completas
df[df.duplicated()]

# Verificar duplicidade por chave de negócio (ex: pedido_id)
df.duplicated(subset=['pedido_id']).sum()

# Visualizar possíveis pedidos repetidos
df[df.duplicated(subset=['pedido_id'], keep=False)]
```

#2
A duplicidade é nociva apenas quando ocorre de forma completa (linhas idênticas), pois distorce os KPIs. Já duplicidades parciais, como múltiplos registros por pedido_id, podem ser legítimas e não devem ser removidas sem análise.

```python
#Remover apenas duplicidades totais
df = df.drop_duplicates().copy()
```

Duplicatas completas representam erro claro de dados por terem a mesma linha repetida, logo removê-las não causa perda de informação real evitando a  distorção direta nos KPIs.




A estratégia escolhida foi remover apenas duplicidades totais, pois elas representam erros claros de registro e impactam diretamente os KPIs, como receita, lucro e quantidade, gerando valores inflados.

Essa abordagem garante a integridade dos dados ao eliminar inconsistências, preserva registros válidos — como múltiplos itens de um mesmo pedido — e mantém a confiabilidade dos indicadores, evitando distorções nas análises.

## 8. Subindo de nível — Criação de variáveis derivadas

Depois da limpeza, é hora de enriquecer a base com variáveis úteis para análise. fileciteturn2file0

### Tarefa
Crie, se possível:
- `ano`
- `mes`
- `ano_mes`
- `margem_lucro`

### Atenção
A criação de `margem_lucro` exige cuidado com divisões por zero e possíveis valores infinitos.


```python
# Crie aqui as variáveis derivadas
import numpy as np

df['data'] = pd.to_datetime(df['data'], errors='coerce')

df['ano'] = df['data'].dt.year

df['mes'] = df['data'].dt.month

#(formato YYYY-MM)
df['ano_mes'] = df['data'].dt.to_period('M').astype(str)
```

```python
# Calcular margem de lucro
df['margem_lucro'] = df['lucro'] / df['receita']

df.loc[df['receita'] == 0, 'margem_lucro'] = np.nan

# Substituir infinitos por NaN
df['margem_lucro'] = df['margem_lucro'].replace([np.inf, -np.inf], np.nan)
```

```python
#verificando
df[['receita', 'lucro', 'margem_lucro']].head()
df['margem_lucro'].describe()
```

### Reflexão técnica
Explique:
1. O que pode acontecer ao calcular margem de lucro quando a receita é zero?

Ao calcular a margem de lucro (lucro / receita), quando a receita é zero, ocorre uma divisão por zero. Isso pode gerar valores infinitos (inf), indefinidos (NaN) ou até erros, dependendo do contexto.

2. Como você decidiu tratar esse caso?
Para tratar esse caso, optei por substituir essas situações por NaN.
Ao manter como NaN, esses registros podem ser ignorados em cálculos estatísticos ou tratados de forma específica posteriormente, preservando a integridade dos indicadores.
